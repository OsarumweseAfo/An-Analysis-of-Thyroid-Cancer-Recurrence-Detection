```python
#Plotting the histogram of Recurred Thyroid disease
plt.figure(figsize=(6, 4), facecolor='white')
plt.hist(data['Recurred'], bins=25, color='skyblue', edgecolor='black')
plt.title('Histogram of Recurred Thyroid Disease')
plt.xlabel('Recurred')
plt.ylabel('Frequency')
```

The  histogram represents the frequency distribution of a categorical variable showing that most patients did not experience a recurrence of thyroid disease, while a smaller number did.

```python
plt.figure(figsize=(6, 4), facecolor='white')
plt.hist(data['Age'], bins=25, color='skyblue', edgecolor='black')
plt.title('Histogram of Age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

The ages range from approximately 15 to 80. The data is right-skewed, indicating that fewer older individuals are present.

```python
#plotting for more emphasis
sns.boxplot(x=data['Age'])
```

```python
#Overlayed Histogram of Age Distribution by Reoccurrence.
sns.histplot(data[data['Recurred'] == 'Yes']['Age'], bins=30, kde=True, color='red', label='Thyroid Yes', alpha=0.6)
sns.histplot(data[data['Recurred'] == 'No']['Age'], bins=30, kde=True, color='green', label='Thyroid No', alpha=0.6)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution by Thyroid Re-occurrence')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

The chart above indicates that Most individuals without thyroid recurrence fall between the 20 - 50 age group, the presence of thyroid recurrence is spread evenly but with higher cases in older age groups from 55 and above.
**In summary, thyroid recurrence is more frequent in older individuals**.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
#Initialize LabelEncoder
label_encoder = LabelEncoder()
```

```python
# List of categorical columns
categorical_cols = ['Gender', 'Smoking', 'SmokingHistory', 'RadiotherapyHistory',
                    'ThyroidFunction', 'PhysicalExamination', 'Adenopathy',
                    'Pathology', 'Focality', 'Risk','T','N','M', 'Stage', 'Response', 'Recurred']  # Include Target Column
```

```python
# Apply Label Encoding
for col in categorical_cols:
    data[col] = label_encoder.fit_transform(data[col])
```

```python
data.columns
```

```python
print("Encoded Data:\n", data.head())
```

```python
# Define features (X) and target variable (y)
X = data[['Age', 'Gender', 'Smoking', 'SmokingHistory', 'RadiotherapyHistory',
       'ThyroidFunction', 'PhysicalExamination', 'Adenopathy', 'Pathology',
       'Focality', 'Risk', 'T', 'N', 'M', 'Stage', 'Response'
]]
y = data['Recurred']  
```

```python
X = data.drop(columns=['Recurred'],axis =1)  # Features
y = data['Recurred']  # Target variable
```

# **Hyperparameter Tuning**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Define RandomForest Model
rf = RandomForestClassifier(random_state=42)
```

```python
# Define hyperparameters
param_dist = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'] 
}
```

```python
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=20,  # Number of random combinations to try
    cv=5,  # 5-fold cross-validation
    scoring='accuracy',
    n_jobs=-1,  # Use all processors
    random_state=42
)
```

```python
# Fit the model on training data
random_search.fit(X_train, y_train)
```

```python
print("Best Parameters:", random_search.best_params_)
```

```python
# Get the best model
best_rf = random_search.best_estimator_
```

```python
# Evaluate on test data
accuracy = best_rf.score(X_test, y_test)
print(f"Optimized Model Accuracy: {accuracy:.4f}")
```
